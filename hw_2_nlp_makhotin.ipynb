### d)

Вот первые 100 биграмм отсортированных по частоте:

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(2,2))
X = vectorizer.fit_transform([''.join(emails)]).toarray()

sorted_bigramms = np.argsort(X[0])[::-1]
bigramms = vectorizer.get_feature_names()
for i in sorted_bigramms[:100]:
    print(bigramms[i], X[0][i]/np.sum(X[0]))
```

В основном это биграммы, которые часто встречаются в любом тексте (связки в предложениях, частые слова с артиклями), но есть и биграммы типа "state gov", "white house", "private residence" etc. которые могут сказать какой тематикой объеденены сообщения.

### e)

Коллокации из двух слов	по PMI, первые 50. Видно что они не очень информативны.

```python
from nltk.collocations import *
from nltk.tokenize import word_tokenize

bigram_measures = nltk.collocations.BigramAssocMeasures()
# Ngrams with 'creature' as a member

## Bigrams
finder = BigramCollocationFinder.from_words(word_tokenize(''.join(emails)))
# only bigrams that appear 3+ times
finder.apply_freq_filter(3)
# return the 10 n-grams with the highest PMI
print(finder.nbest(bigram_measures.likelihood_ratio, 50))
```

Удалим стоп-слова и знаки пунктуации, получится гораздо лучше:

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
import string

stop = stopwords.words('english') + list(string.punctuation)
words = [i for i in word_tokenize(''.join(emails)) if i not in stop]

finder = BigramCollocationFinder.from_words(words)
# only bigrams that appear 3+ times
finder.apply_freq_filter(3)
# return the 10 n-grams with the highest PMI
print(finder.nbest(bigram_measures.likelihood_ratio, 50))
```

### f, g)

Применим к сообщениям tfidf дескриптор, и попробуем раздлить на 3 кластера при помощи k-means. Визуализируем каждый кластер как 10 сообщений из него, можно взглянуть на нихи определить объеденены ли они какой-либо тематикой.

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

def visualize_clusters(X, model):
    preds = model.fit_predict(X)
    for i in range(len(set(model.labels_)) - 1):
        for j in range(min(len(emails[preds == i]), 10)):
            print(str(j) + ') ' + emails[preds == i][j])
        print('-'*111  +'\n')
    
    n = len(set(model.labels_))
    if -1 in model.labels_:
        n -= 1
    for i in range(n):
        print('letters in cluster #' + str(i) + ' :' + str(len(emails[preds == i])))
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(emails)
model = KMeans(n_clusters=3, random_state=147)
visualize_clusters(X, model)
```

Видно, что в один кластер попали короткие сообщения, которые содержат "Ok", во второй короткие, которые содержат "FYI", в третий все остальное.

Далее попробуем ограничить document frequency сверху и снизу в tfidf, и сделаем то же самое.

```python
vectorizer = TfidfVectorizer(max_df=500, min_df=30, stop_words='english')
X = vectorizer.fit_transform(emails)
model = KMeans(n_clusters=3, random_state=147)
visualize_clusters(X, model)
```

Получили все тот же кластер с "Ok", более интересный кластер с расписаниями, и кластер со всем остальным.

Попробуем разбить на 5 кластеров.

```python
vectorizer = TfidfVectorizer(max_df=500, min_df=30, stop_words='english')
X = vectorizer.fit_transform(emails)
model = KMeans(n_clusters=5, random_state=147)
visualize_clusters(X, model)
```

Попробуем применить PCA, возьмем 1000 компонент и разделим на 3 кластера.

```python
from sklearn.decomposition import TruncatedSVD

vectorizer = TfidfVectorizer(max_df=500, min_df=30, stop_words='english')
X = vectorizer.fit_transform(emails)
svd = TruncatedSVD(n_components=1000, random_state=228)
X = svd.fit_transform(X)
model = KMeans(n_clusters=3, random_state=147)
visualize_clusters(X, model)
```

Получилось то же самое, что и без PCA.

### h)

Имеем, что k-means всегда выделяет один большой неинформативный кластер, а остальные получаются маленькие, но информативные. Так же с k-means существует проблема, что количество кластеров это гиперпараметр, который нужно подбирать. Попробуем убить эти две проблемы агломеративным методом кластеризации DBSCAN, который подбирает автоматически количество кластеров и не кластеризует вообще элементы выборки, которые не входят ни в какой из кластеров.

```python
from sklearn.cluster import DBSCAN

model = DBSCAN()
visualize_clusters(X, model)
```

Получилось 39 кластеров, но они все маленькие, состоят в среднем из нескольких десятков сообщений, но каждый кластер, если посмотреть, получился интерпретируемым.

### i)

В моем случае интерпретируемость можно проверить и самому, так как кластеры получились небольшие. Но в случае если кластеры были бы достаточно большие можно было бы проверить интерпретируемость с помощью  асессоров следующим образом. Например, давать асессору несколько писем из кластера, определенного алгоритмом и спрашивать, можно ли эти письма отнести к одному кластеру. Если мы будем слишком часто получать отрицательные ответы, значит, понятное дело, кластеризация получилась неинтерпритируемая.
